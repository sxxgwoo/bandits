```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#slot machine 당기는 횟수
T = 1000

#slot machine들의 expected reward
##slot machine 2개
##1번째 당기면 보상 기댓값 3, 2번째 당기면 보상 기댓값 5
##assume: r_i(t) = mu_i + epsilon, epsilon = N(0,1) 정규분포 independence
N=2

#mu = [mu1,mu2, ... ]
mu = [3,5]

cum_obs_regret_random = [0]

for t in range(T):
    # an algorithm selects item i (denoted by a(t)).
    arm_random = t % 2

    # the algorithm observes reward for the item (denoted by r_{a(t)}(t))
    rwd_random = mu[arm_random] + np.random.normal()

    # if t % 100 == 0 :
    #     print(t)
    #     print(arm_random)
    #     print(rwd_random)

    # (for validation purpose) calculate cumulative regret
    # def. regret(t) = (expected reward from optimal arm) - (expected reward from selected arm)
    # def. cumulative regret = sum_{t=1}^T regret)(t)
    # calcuate: cumulative observed regret (true optimal arm을 안다고 가정)
    # def. obs_regret(t) = (expected reward from optimal arm) - (observed reward from selected arm)
    # def. cumulative observed regret = sum_{t=1}^T obs_regret(t)
    obs_regret_random = max(mu) - rwd_random
    cum_obs_regret_random.append(cum_obs_regret_random[t]+obs_regret_random)

# cumulative sum 값 확인
# print(cum_obs_regret_random)

# plot으로 확인
## t=0 일때 2.xxxx
## t=1 일때 0.xxx
## 따라서 짝수일때 대략 (T/2)*2 이므로 T 선형적으로 그래프가 그려짐
plt.plot(cum_obs_regret_random)
```
